```python
"""
OOP Practice
- Class inheritance
- Basic object modeling
- Device abstraction
"""

class Device:
    count = 0

    def __init__(self, ip, mac, name, status="unknown"):
        self.ip = ip
        self.mac_address = mac
        self.name = name
        self.status = status

        Device.count += 1


class TV(Device):
    def turn_on(self):
        self.status = "active"
        print(f"{self.name} is now ON")

    def turn_off(self):
        self.status = "inactive"
        print(f"{self.name} is now OFF")


# Example usage
tv1 = TV("192.168.1.10", "AA:BB:CC:DD:EE:FF", "Living Room TV")

tv1.turn_on()
tv1.turn_off()

print("\nTotal devices:", Device.count)
```
