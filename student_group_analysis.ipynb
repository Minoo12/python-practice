```python
"""
This program:
Collects student data (age, height, weight)
Computes averages for two classes
Compares the classes based on height and weight
"""

class StudentGroup:
    def __init__(self):
        self.ages = []
        self.heights = []
        self.weights = []

    def input_data(self, n):
        for _ in range(n):
            age = float(input("Enter age: "))
            height = float(input("Enter height: "))
            weight = float(input("Enter weight: "))

            self.ages.append(age)
            self.heights.append(height)
            self.weights.append(weight)

    def average(self, values):
        return sum(values) / len(values)


# Create two groups
class_a = StudentGroup()
class_b = StudentGroup()

# Input data for Class A
n1 = int(input("Enter number of students in Class A: "))
class_a.input_data(n1)

# Input data for Class B
n2 = int(input("Enter number of students in Class B: "))
class_b.input_data(n2)

# Compute averages
avg_age_a = class_a.average(class_a.ages)
avg_height_a = class_a.average(class_a.heights)
avg_weight_a = class_a.average(class_a.weights)

avg_age_b = class_b.average(class_b.ages)
avg_height_b = class_b.average(class_b.heights)
avg_weight_b = class_b.average(class_b.weights)

# Display results
print("\nClass A averages:", avg_age_a, avg_height_a, avg_weight_a)
print("Class B averages:", avg_age_b, avg_height_b, avg_weight_b)

# Comparison logic
if avg_height_a > avg_height_b:
    print("\nResult: Class A has higher average height")
elif avg_height_b > avg_height_a:
    print("\nResult: Class B has higher average height")
else:
    if avg_weight_a < avg_weight_b:
        print("\nResult: Class A is considered better (lower weight)")
    elif avg_weight_b < avg_weight_a:
        print("\nResult: Class B is considered better (lower weight)")
    else:
        print("\nResult: Both classes are equal")
```
