```python
"""
Mastering Lambda Functions, Map, Filter & Sorting in Python
- Lambda expressions
- map(), filter(), and sorted()
- sorting with keys
"""

# ==================== 1. Sorting with Lambda Key ====================

print("=== Sorting Tuples by Specific Element ===")
a = [(3, 4), (7, 1), (5, 9), (2, 2)]
print("Original:", a)

# Sort by the second element of each tuple 
a.sort(key=lambda x: x[1])
print("Sorted by 2nd element:", a)
print()


# ==================== 2. Transforming Data with map() ====================

print("=== Transforming Lists with map() ===")
my_list = [1, 5, 6, 8, 10, 11]
doubled = list(map(lambda x: x * 2, my_list))
print("Original:", my_list)
print("Doubled:", doubled)
print()


# ==================== 3. Conditional Transformation ====================

print("=== Conditional Mapping ===")
numbers = [10, 11, 6, 100, 7, 0.5]
size_label = list(map(lambda x: "big" if x > 10 else "small", numbers))
print("Numbers:", numbers)
print("Size labels:", size_label)
print()


# ==================== 4. Even/Odd Classification ====================

print("=== Even/Odd Check with map() ===")
mylista = [2, 3, 5, 8, 11, 14, 17, 102, 44]
parity = list(map(lambda x: 'Yes' if x % 2 == 1 else 'No', mylista))
print("Numbers:", mylista)
print("Odd? (Yes/No):", parity)
print()


# ==================== 5. Multi-element Sorting ====================

print("=== Sorting by Third Element ===")
array = [(1, 4, 5), (3, 2, 7), (8, 3, 6), (9, 2, 3)]
print("Original:", array)

array.sort(key=lambda a: a[2])
print("Sorted by 3rd element:", array)
print()


# ==================== 6. String Processing & Filtering ====================

print("=== String Processing & Filtering ===")
colors = ['yellow', 'red', 'blue', 'red', 'yellow', 'red', 'blue', 'purple']
colors.sort()

# Replace 'red' with 'color' while keeping others
processed_colors = list(map(lambda x: 'color' if x == 'red' else x, colors))

print("Sorted colors:", colors)
print("Processed:", processed_colors)

# Filter only 'red'
red_only = list(filter(lambda x: x == 'red', colors))
print("Red only:", red_only)
print()


# ==================== 7. Chaining map() and filter() ====================

print("=== Chaining map() + filter() ===")
numbers = [2, 15, 26, 8, 11, 14, 17, 102, 44]

# Get last digit
last_digits = list(map(lambda x: x % 10, numbers))

# Filter last digits <= 4
small_digits = list(filter(lambda x: x <= 4, last_digits))

print("Original:", numbers)
print("Last digits:", last_digits)
print("Last digits <= 4:", small_digits)
print()


# ==================== 8. Another Chained Example ====================

print("=== Another Chained Example (mod 5) ===")
mod5 = list(map(lambda x: x % 5, numbers))
filtered_mod5 = list(filter(lambda x: x < 4, mod5))

print("Numbers mod 5:", mod5)
print("Filtered (< 4):", filtered_mod5)
```
