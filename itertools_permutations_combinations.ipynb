```python
"""
Generating permutations (ordered arrangements)
Generating combinations (unordered selections)
"""

import itertools

def demonstrate_permutations():
    """
    Demonstrate permutations 
    """
    print(" PERMUTATIONS")
    print("=" * 50)
    
    numbers = [1, 2, 3]
    print(f"Original list: {numbers}\n")
    
    # Permutations of 2 items out of 3
    perm2 = list(itertools.permutations(numbers, 2))
    print(f"Permutations of 2 items: ({len(perm2)} total)")
    for p in perm2:
        print(f"  {p}")
    
    print()
    
    # All permutations (r = len)
    perm_all = list(itertools.permutations(numbers))
    print(f"All permutations: ({len(perm_all)} total)")
    for p in perm_all:
        print(f"  {p}")
    
    print("-" * 50)


def demonstrate_combinations():
    """
    Demonstrate combinations 
    """
    print("COMBINATIONS")
    print("=" * 50)
    
    numbers = [1, 2, 3]
    print(f"Original list: {numbers}\n")
    
    # Combinations of 2 items out of 3
    comb2 = list(itertools.combinations(numbers, 2))
    print(f"Combinations of 2 items: ({len(comb2)} total)")
    for c in comb2:
        print(f"  {c}")
    
    # Combinations with repetition
    comb_with_rep = list(itertools.combinations_with_replacement(numbers, 2))
    print(f"\nCombinations with repetition: ({len(comb_with_rep)} total)")
    for c in comb_with_rep:
        print(f"  {c}")
    
    print("-" * 50)


def demonstrate_real_world_example():
    """ Team selection & scheduling  """
    print(" REAL-WORLD EXAMPLE")
    print("=" * 50)
    
    players = ['Alice', 'Bob', 'Charlie', 'Diana']
    
    print("Players:", players)
    print("\nPossible pairs for a doubles match (Combinations):")
    for pair in itertools.combinations(players, 2):
        print(f"  {pair}")
    
    print("\nPossible ordered starting lineups of 3 players (Permutations):")
    for lineup in list(itertools.permutations(players, 3))[:6]:  # Show only first 6
        print(f"  {lineup}")
    print(f"  ... and {len(list(itertools.permutations(players, 3)))} more!")


# ==================== Main ====================

def main():
    print("🧠 itertools: Permutations & Combinations Demo\n")
    
    demonstrate_permutations()
    demonstrate_combinations()
    demonstrate_real_world_example()
    
    print("\n" + "=" * 60)
    print("Key Difference:")
    print("• Permutations → Order matters     (AB ≠ BA)")
    print("• Combinations → Order doesn't matter (AB = BA)")
    print("=" * 60)


if __name__ == "__main__":
    main()
```
