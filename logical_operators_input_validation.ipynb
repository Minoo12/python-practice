```python
"""
     Logical operators (and, or)
"""

# Input values
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

print("\nValues entered:")
print(a, b, c, sep="\n")

# Logical condition 1
is_sorted = (a < b) and (b <= c)
print(f"\nIs a < b <= c : {is_sorted}")

# Additional input
d = int(input("\nEnter fourth number: "))

# Logical condition 2
condition = (a > b) or (d > b)
print(f"Is a > b or d > b : {condition}")
```
