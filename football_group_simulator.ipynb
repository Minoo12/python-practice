```python
"""
Football Group Stage Simulator
Based on a programming exercise from an online course.
Implemented and extended by: Minoo Sh
"""


teams = ["Iran", "Morocco", "Spain", "Portugal"]

matches = [
    ("Iran", "Morocco"),
    ("Iran", "Spain"),
    ("Iran", "Portugal"),
    ("Morocco", "Spain"),
    ("Morocco", "Portugal"),
    ("Spain", "Portugal")
]

# Points dictionary
points = {team: 0 for team in teams}


# ==================== Helper Functions ====================

def play_match(team1, team2):

    print(f"\n=== Match: {team1} vs {team2} ===")
    
    while True:
        try:
            goals1 = int(input(f"{team1} goals: "))
            goals2 = int(input(f"{team2} goals: "))
            if goals1 >= 0 and goals2 >= 0:
                break
            print("Goals cannot be negative. Try again.")
        except ValueError:
            print("Please enter valid numbers.")

    print("-" * 50)

    if goals1 > goals2:
        points[team1] += 3
        print(f"✅ Winner: {team1} {goals1} - {goals2} {team2}")
    elif goals1 < goals2:
        points[team2] += 3
        print(f"✅ Winner: {team2} {goals2} - {goals1} {team1}")
    else:
        points[team1] += 1
        points[team2] += 1
        print(f"🤝 Draw! {team1} {goals1} - {goals2} {team2}")

    return goals1, goals2


def display_standings():
    """Display current or final standings sorted by points."""
    print("\n" + "=" * 55)
    print("🏆 CURRENT GROUP STANDINGS 🏆")
    print("=" * 55)
    
    # Sort teams by points (descending)
    sorted_teams = sorted(teams, key=lambda t: points[t], reverse=True)
    
    print(f"{'Position':<10} {'Team':<15} {'Points':<10}")
    print("-" * 55)
    for pos, team in enumerate(sorted_teams, 1):
        print(f"{pos:<10} {team:<15} {points[team]:<10}")
    print("=" * 55)


# ==================== Main  ====================

def main():
    print("⚽ Welcome to the Football Group Stage Simulator ⚽\n")
    print("Teams:", ", ".join(teams))
    print("=" * 60)

    # Play all matches
    for i, (team1, team2) in enumerate(matches, 1):
        print(f"\nMatch {i} of {len(matches)}")
        play_match(team1, team2)
        display_standings()  # Show standings after each match

    # Final Results
    print("\n" + "=" * 60)
    print("🏆 FINAL GROUP STANDINGS 🏆")
    print("=" * 60)
    
    sorted_teams = sorted(teams, key=lambda t: points[t], reverse=True)
    
    for pos, team in enumerate(sorted_teams, 1):
        print(f"{pos}. {team:<15} — {points[team]} points")


if __name__ == "__main__":
    main()


    
```
