```python
"""
To-Do List Manager 
"""

import csv
import os
from datetime import datetime
from typing import List


class Task:
    def __init__(self, name: str, description: str = "", priority: str = "medium"):
        self.name = name.strip()
        self.description = description.strip()
        self.priority = priority.lower()
        self.created_at = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        if self.priority not in ["high", "medium", "low"]:
            self.priority = "medium"

    def __str__(self) -> str:
        labels = {"high": "[HIGH]", "medium": "[MEDIUM]", "low": "[LOW]"}
        label = labels.get(self.priority, "[MEDIUM]")
        desc = f" - {self.description}" if self.description else ""
        return f"{label} {self.name}{desc}"

    def to_dict(self) -> dict:
        return {
            "name": self.name,
            "description": self.description,
            "priority": self.priority,
            "created_at": self.created_at
        }

    @classmethod
    def from_dict(cls, data: dict) -> 'Task':
        task = cls(data.get("name", ""), data.get("description", ""), data.get("priority", "medium"))
        task.created_at = data.get("created_at", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
        return task


class ToDoList:
    def __init__(self, filename: str = "todo_list.csv"):
        self.tasks: List[Task] = []
        self.filename = filename
        self.load_from_csv()

    def add_task(self, name: str, description: str = "", priority: str = "medium"):
        if not name.strip():
            print(" Task name cannot be empty!")
            return False
        
        task = Task(name, description, priority)
        self.tasks.append(task)
        self.save_to_csv()
        print(f"Task added: {task}")
        return True

    def remove_task(self, index: int):
        if 0 <= index < len(self.tasks):
            removed = self.tasks.pop(index)
            self.save_to_csv()
            print(f" Removed: {removed.name}")
            return True
        print("Invalid task number!")
        return False

    def show_all_tasks(self):
        if not self.tasks:
            print("\n Your to-do list is currently empty.")
            return
        
        print("\n" + "=" * 75)
        print(" YOUR TO-DO LIST")
        print("=" * 75)
        
        for i, task in enumerate(self.tasks):
            print(f"{i:2d}. {task}")
        
        print("=" * 75)
        print(f"Total tasks: {len(self.tasks)}")
        print("=" * 75)

    def save_to_csv(self):
        try:
            with open(self.filename, 'w', newline='', encoding='utf-8') as f:
                writer = csv.DictWriter(f, fieldnames=['name', 'description', 'priority', 'created_at'])
                writer.writeheader()
                for task in self.tasks:
                    writer.writerow(task.to_dict())
        except Exception as e:
            print(f"❌ Save error: {e}")

    def load_from_csv(self):
        if not os.path.exists(self.filename):
            return
        
        try:
            with open(self.filename, 'r', encoding='utf-8') as f:
                reader = csv.DictReader(f)
                self.tasks = [Task.from_dict(row) for row in reader]
            print(f"📂 Loaded {len(self.tasks)} task(s) from file.")
        except Exception:
            self.tasks = []


# ==================== Menu ====================

def display_menu():
    print("\n" + "=" * 60)
    print("TO-DO LIST")
    print("=" * 60)
    print("1.Add New Task")
    print("2.Remove Task")
    print("3.View To-Do List")
    print("4.Save")
    print("5.Exit")
    print("=" * 60)


def get_priority() -> str:
    print("\nPriority:")
    print("1. High ")
    print("2. Medium ")
    print("3. Low")
    while True:
        ch = input("Choose (1-3): ").strip()
        if ch == "1": return "high"
        if ch == "2": return "medium"
        if ch == "3": return "low"
        print("Invalid choice.")


# ==================== Main ====================

def main():
    print("\n Welcome to To-Do List Manager")
    todo = ToDoList()

    while True:
        display_menu()
        choice = input("\nEnter choice (1-5): ").strip()

        if choice == "1":
            print("\n--- Add New Task ---")
            name = input("Task name: ").strip()
            desc = input("Description (optional): ").strip()
            priority = get_priority()
            
            todo.add_task(name, desc, priority)
            todo.show_all_tasks()          # ← Auto show after adding

        elif choice == "2":
            todo.show_all_tasks()
            if todo.tasks:
                try:
                    idx = int(input("\nEnter task number to remove: "))
                    todo.remove_task(idx)
                    todo.show_all_tasks()   # ← Show updated list
                except ValueError:
                    print("Please enter a valid number!")

        elif choice == "3":
            todo.show_all_tasks()

        elif choice == "4":
            todo.save_to_csv()
            print("List saved successfully!")

        elif choice == "5":
            print("\n Goodbye! Your tasks are saved.")
            break

        else:
            print("Invalid choice! Please select 1-5.")


if __name__ == "__main__":
    main()
```
