```python
"""
Set Operations in Python
"""

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Add a new element
set_a.add(17)

# Perform set operations
union = set_a.union(set_b)
intersection = set_a.intersection(set_b)

# Display results
print(f"Set A: {set_a}")
print(f"Set B: {set_b}")
print(f"Union: {union}")
print(f"Intersection: {intersection}")
```
