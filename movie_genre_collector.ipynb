```python
# ==================== Configuration ====================

favorites = ["Horror", "Romance", "Comedy", "History", "Adventure", "Action"]

# ==================== Main Program ====================

def main():
    print("🎬 Welcome to the Movie Genre Collector 🎬\n")
    print("Available genres:", ", ".join(favorites))
    print("=" * 60)

    # Get number of entries
    while True:
        try:
            num_people = int(input("\nHow many people do you want to add? "))
            if num_people > 0:
                break
            print("Please enter a positive number.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    names = []  # 2D list to store [name, genre] pairs

    # Collect data
    for i in range(num_people):
        print(f"\n--- Person {i+1} ---")
        
        my_name = input("Enter name: ").strip()
        
        # Get and validate genre
        your_genre = input("Enter favorite genre: ").strip().title()
        
        if your_genre in favorites:
            print(f"✅ {my_name} added successfully!")
        else:
            print(f"❌ '{your_genre}' is not available.")
            print("Available genres:", favorites)
            while your_genre not in favorites:
                your_genre = input("Please choose a valid genre: ").strip().title()
            print(f"✅ {my_name} added with {your_genre}!")

            
        names.append([my_name, your_genre])

    # ==================== Display Results ====================

    print("\n" + "=" * 60)
    print("ALL ENTRIES")
    print("=" * 60)
    
    for person in names:
        print(f"• {person[0]:<20} likes {person[1]}")

    # Display as formatted table
    print("\n" + "=" * 60)
    print("2D MATRIX - FORMATTED TABLE")
    print("=" * 60)
    print(f"{'Index':<6} {'Name':<20} {'Favorite Genre':<20}")
    print("-" * 60)
    
    for i, row in enumerate(names):
        print(f"{i:<6} {row[0]:<20} {row[1]:<20}")

    # Raw 2D list
    print("\n" + "=" * 60)
    print("2D LIST (Matrix):")
    print("=" * 60)
    for row in names:
        print(row)


# Run the program
if __name__ == "__main__":
    main()
```
