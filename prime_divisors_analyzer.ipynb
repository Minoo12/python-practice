```python
"""
This code:
Calculates the number of distinct prime divisors for each
Finds the number with the maximum count of prime divisors
"""

# ==================== Helper Functions ====================

def is_prime(num: int) -> bool:
    """Check if a number is prime."""
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True


def count_distinct_prime_divisors(n: int) -> int:

    if n <= 1:
        return 0
    
    count = 0
    temp = n
    
    # Handle factor 2 separately
    if temp % 2 == 0:
        count += 1
        while temp % 2 == 0:
            temp //= 2
    
    # Check odd factors
    i = 3
    while i * i <= temp:
        if temp % i == 0:
            count += 1
            while temp % i == 0:
                temp //= i
        i += 2
    
    # If remaining number is a prime > 2
    if temp > 2:
        count += 1
    
    return count


# ==================== Main Program ====================

def main():
    print("Prime Divisors Analyzer\n")
    print("Enter 10 integers (positive or negative - negatives will be treated as absolute value)\n")
    
    numbers = []
    
    # Input collection with validation
    for i in range(10):
        while True:
            try:
                num = int(input(f"Number {i+1}: "))
                numbers.append(abs(num))  # Use absolute value for negative numbers
                break
            except ValueError:
                print("Please enter a valid integer.")

    # Analyze each number
    max_count = -1
    result_number = None
    results = []  # Store all results for detailed display

    for num in numbers:
        count = count_distinct_prime_divisors(num)
        results.append((num, count))
        
        # Update max: higher count, or same count but larger number
        if count > max_count or (count == max_count and num > result_number):
            max_count = count
            result_number = num

    # ==================== Display Results ====================

    print("\n" + "=" * 60)
    print("ANALYSIS RESULTS")
    print("=" * 60)
    
    print(f"{'Number':<12} {'Prime Divisors Count':<25}")
    print("-" * 60)
    for num, count in results:
        print(f"{num:<12} {count:<25}")
    
    print("\n" + "=" * 60)
    print("FINAL RESULT")
    print("=" * 60)
    print(f"Number with the most distinct prime divisors : {result_number}")
    print(f"Count of distinct prime divisors              : {max_count}")
    print("=" * 60)


if __name__ == "__main__":
    main()
```
