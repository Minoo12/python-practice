```python
class Computer:
    def __init__(self, ram, hard, cpu):
        self.ram = ram
        self.hard = hard
        self.cpu = cpu

    def value(self):
        return self.ram + self.hard + self.cpu


class Laptop(Computer):
    def __init__(self, ram, hard, cpu, size):
        super().__init__(ram, hard, cpu)
        self.size = size

    def value(self):
        return super().value() + self.size


pc = Computer(12, 2, 4)
laptop = Laptop(12, 2, 4, 13)

print(pc.value())
print(laptop.value())
```
