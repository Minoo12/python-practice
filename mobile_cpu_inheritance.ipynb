```python
class Mobile:
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

    def get_name(self):
        return self.name


class CPU(Mobile):
    def get_name(self):
        print(f"This mobile has HighTech {self.name} CPU")


device = CPU("Intel", "iphone")
device.get_name()
```
